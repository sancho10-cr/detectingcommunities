```python
import numpy as np
import networkx as nx
from sklearn.cluster import KMeans
from scipy.io import mmread

# Load the MTX file: Email, the university network as a sparse adjacency matrix
#adj_matrix = mmread("../inputs/email/email.mtx").tocsr()

# Convert the sparse adjacency matrix to a NetworkX graph
#G = nx.from_scipy_sparse_matrix(adj_matrix)
# Load the Books about US politics networks dataset
#G = nx.read_gml('../inputs/polbooks/polbooks.gml')
# Load the US power grid network
#G = nx.read_gml('../inputs/power/power.gml', label = 'id')

# Load the Bottlenose Dolphins network dataset
G = nx.read_gml('../inputs/dolphins/dolphins.gml')

# Convert the graph to a numpy array for k-means clustering
node_features = nx.to_numpy_array(G)

# Apply k-means clustering
k = 2  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(node_features)
labels = kmeans.labels_

# Generate the true labels based on the k-means clustering results
true_labels = {node: label for node, label in enumerate(labels)}

_true=[]
# Print the true labels for each node
for node, true_label in true_labels.items():
    _true.append(true_label)
    print(f"Node {node}: True Label {true_label}")
print(_true)
```

```python
import networkx as nx
import community

# Load the Bottlenose Dolphins network dataset
#G = nx.read_gml('../inputs/dolphins/dolphins.gml')

# Load the US power grid network
G = nx.read_gml('../inputs/power/power.gml', label = 'id')

# Apply the Louvain algorithm for community detection
partition = community.best_partition(G)
true_label = []
# Print the true labels (community assignments) for each dolphin
for dolphin, label in partition.items():
    true_label.append(label)
    print(f"Dolphin {dolphin}: Community {label}")
print(true_label)
```

```python
import networkx as nx
from networkx.algorithms import community

# Load the Bottlenose Dolphins network dataset
G = nx.read_gml('../inputs/dolphins/dolphins.gml')

# Perform community detection using the Girvan-Newman algorithm
communities_generator = community.girvan_newman(G)
communities = next(communities_generator)

# Divide the nodes into two communities
community_1 = communities[0]
community_2 = communities[1]

# Print the nodes in each community
print("Community 1:")
print(community_1)
print("\nCommunity 2:")
print(community_2)
```
